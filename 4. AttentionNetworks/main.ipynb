### Why attention network?

RNN: Struggle with long-term dependencies due to the vanishing gradient problem, even with advanced variants like LSTMs or GRUs.
Process inputs sequentially, making them slow 

CNN: Can capture local dependencies effectively but require deep architectures or large receptive fields to model long-range relationships. 
Are parallelizable but might still require extensive layers to process sequences fully.

### Main Motive of Attention networks..

-> long-range dependencies 

-> focusing on relevant parts.
