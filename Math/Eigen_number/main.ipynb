```python
def factorial(n):
    return 1 if n == 0 else n*factorial(n-1)

def exp(x):
    return sum([
        x**n/factorial(n)
        for n in range(150)
    ])

exp(1)
```

```python
exp(5)
```

```python
exp(1)**5
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 2 * np.pi, 100)
plt.plot([exp(complex(0, t)).real for t in x], [exp(complex(0, t)).imag for t in x])
plt.xlabel('Real part')
plt.ylabel('Imaginary part')
plt.title('Plot of exp(i*x) for x in range(100)')
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
