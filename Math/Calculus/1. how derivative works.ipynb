```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2
def derviate_2_power_x(x):
    h = 0.0001
    return (f(x) - f(x-h)) / h

x = np.arange(0, 10)   
plt.plot(x, f(x), label='2^x')
plt.plot(x, derviate_2_power_x(x),  label='Actual Derivative 2^x', color='red')  
plt.title('Derivative of 2^x')
plt.xlabel('x')
plt.ylabel('f\'(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.exp(x)

def derviate_e_power_x(x):
    h = 0.0001
    return (f(x) - f(x-h)) / h

x = np.arange(0, 10, 0.1)

y = np.array([f(xi) for xi in x])
dy = np.array([derviate_e_power_x(xi) for xi in x])

fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].plot(x, y, label='e^x', color='blue')
ax[0].set_xlabel('x')   # <- corrected
ax[0].set_ylabel('f(x)')
ax[0].legend()
ax[0].grid(True)

ax[1].plot(x, dy, label="Derivative of e^x", color='red')
ax[1].set_xlabel('x')   # <- corrected
ax[1].set_ylabel("f'(x)")
ax[1].legend()
ax[1].grid(True)

fig.suptitle('e^x and its Derivative')  # <- corrected
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.sin(x)

def derviate_sin_x(x):
    h = 0.0001
    return (f(x) - f(x-h)) / h

x = np.arange(0, 50, 0.01)

y = np.array([f(xi) for xi in x])
dy = np.array([derviate_sin_x(xi) for xi in x])

fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].plot(x, y, label='sin(x)', color='blue')
ax[0].set_xlabel('x')   # <- corrected
ax[0].set_ylabel('f(x)')
ax[0].legend()
ax[0].grid(True)

ax[1].plot(x, dy, label="Derivative of sin(x)-> cos(x)", color='red')
ax[1].set_xlabel('x')   # <- corrected
ax[1].set_ylabel("f'(x)")
ax[1].legend()
ax[1].grid(True)

fig.suptitle('sin(x) and its Derivative')  # <- corrected
plt.show()


# At x=0x=0:

#     f(0)=sin(0)=0f(0)=sin(0)=0

#     f(h)=sin(0.0001)0.0001f(h)=sin(0.0001)0.0001

# So,
# f(0)0(0.0001)0.0001=0.00010.0001=1
# f(0)0.00010(0.0001)=0.00010.0001=1

#  It matches exactly what we expect!
# Derivative of sin(x) at x=0 is cos(0) = 1.
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.cos(x)

def derviate_cos_x(x):
    h = 0.0001
    return (f(x) - f(x-h)) / h

x = np.arange(0, 50, 0.01)

y = np.array([f(xi) for xi in x])
dy = np.array([derviate_cos_x(xi) for xi in x])

fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].plot(x, y, label='cos(x)', color='blue')
ax[0].set_xlabel('x')   # <- corrected
ax[0].set_ylabel('f(x)')
ax[0].legend()
ax[0].grid(True)

ax[1].plot(x, dy, label="Derivative of cos(x)-> -sin(x)", color='red')
ax[1].set_xlabel('x')   # <- corrected
ax[1].set_ylabel("f'(x)")
ax[1].legend()
ax[1].grid(True)

fig.suptitle('cos(x) and its Derivative')  # <- corrected
plt.show()


# At x=0x=0:

#     f(0)=1  

#     f(0h)=cos(0.0001)0.999999995f(0h)=cos(0.0001)0.999999995

#     So:

# dy=10.999999995/0.00010.00005
```
