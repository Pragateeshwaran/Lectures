```python
import torch
import torch.nn as nn
import torchvision
import torch.nn.functional as F
import os   
```

```python
os.path.exists(r'F:\works\A-important\A-neurals\CV\images\Image.jpg')
```

```python
img = torchvision.io.read_image(r'F:\works\A-important\A-neurals\CV\images\Image.jpg')
img = (img) / 255.0  
img.shape
```

```python
class ConvBlock(nn.Module):
    def __init__(self, in_channels=3, out_channels=3):
        super(ConvBlock, self).__init__()
        self.conv = nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1)
        self.relu = nn.ReLU(inplace=True)
    def forward(self, x):
        x = self.conv(x)
        x = self.relu(x)
        return x
```

```python
model = ConvBlock()
res = model(img)
res.shape
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(1, 3, figsize=(12, 3))  # 1 row, 4 columns

for i in range(res.shape[0]):
    axes[i].imshow(res[i, :, :].detach().numpy(), cmap='gray')
    axes[i].axis('off')  # Hide axes

plt.tight_layout()
plt.show()
```

```python

res = model(res) + img

import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(1, 3, figsize=(12, 3))  # 1 row, 4 columns

for i in range(res.shape[0]):
    axes[i].imshow(res[i, :, :].detach().numpy(), cmap='gray')
    axes[i].axis('off')  # Hide axes

plt.tight_layout()
plt.show()
```
